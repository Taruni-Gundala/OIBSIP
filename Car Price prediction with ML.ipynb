```python
#Loading required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
features = ["symboling", "wheelbase", "carlength",
            "carwidth", "carheight", "curbweight",
            "enginesize", "boreratio", "stroke",
            "compressionratio", "horsepower", 
            "peakrpm","citympg", "highwaympg"]
```

```python
x=data[features].values
y=data['price'].values
#Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
#Create an object for DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train,y_train)
#Predict the values
y_pred = regressor.predict(x_test)
print("Predictions : ")
print(y_pred)
```

```python
df = pd.DataFrame({'Actual Values' : y_test , 'Predicted Values' : y_pred})
print(df)
```

```python
#Model Evaluation
error = mean_squared_error(y_test,y_pred)
print('Mean Squared Error: ',error)
```

```python
r2_score = regressor.score(x_test,y_test)
print('R^2 Error: ',r2_score)
```

```python
#Plot the graph for Actual vs Predicted values
plt.figure(figsize=(7,6))
plt.scatter(y_test,y_pred,alpha=0.8,color='blue',label='Prediction')
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color='red',label='Actual')
plt.title('Actual values VS Predicted values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```
